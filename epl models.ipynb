```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
base_path = "C:\\Users\\user\\Downloads"
seasons = [f"{year}/{year+1}" for year in range(2014, 2025)]  # Generate seasons dynamically

# Load datasets and add the 'Season' column
dfs = []
for i, season in enumerate(reversed(seasons)):  # Reverse to match file naming order
    file_name = "E0.csv" if i == 0 else f"E0 ({i}).csv"
    file_path = os.path.join(base_path, file_name)
    
    try:
        df = pd.read_csv(file_path)
        df["Season"] = season
        dfs.append(df)
    except FileNotFoundError:
        print(f"Warning: {file_name} not found. Skipping.")

# Combine all datasets
df_combined = pd.concat(dfs, ignore_index=True)

# Display first few rows
df_combined.head()
```

```python
# Function to parse different date formats
def parse_date(date):
    return pd.to_datetime(date, dayfirst=True, errors='coerce')

# Load and preprocess dataset
df_combined['Date'] = df_combined['Date'].apply(parse_date)
df_combined = df_combined.sort_values(by='Date')

```

```python
df_filtered = df_combined[['Season', 'HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', 'Referee', 'Date', 'HS', 'AS', 'HST', 'AST', 'B365H', 'B365D', 'B365A']]
df_filtered.head()
```

```python
df_filtered = df_filtered.dropna()
df_filtered = df_filtered.drop_duplicates()
```

```python
# Encode FTR (target variable)
df_filtered['FTR'] = df_filtered['FTR'].map({'A': 0, 'D': 1, 'H': 2})

# Encode team names as numerical IDs
df_filtered['HomeTeam'] = df_filtered['HomeTeam'].astype('category').cat.codes
df_filtered['AwayTeam'] = df_filtered['AwayTeam'].astype('category').cat.codes

# Encode Refree as a categorical variable
df_filtered['Referee'] = df_filtered['Referee'].astype('category').cat.codes
```

```python
# Convert Date and Time to datetime
df_filtered['DateTime'] = pd.to_datetime(df_filtered['Date'])

# Extract features
df_filtered['DayOfWeek'] = df_filtered['DateTime'].dt.dayofweek  # Monday=0, Sunday=6
df_filtered['Month'] = df_filtered['DateTime'].dt.month


# Drop original Date and Time columns
df_filtered = df_filtered.drop(columns=['Date'])
df_filtered.head()
```

```python
#### Feature Engineering ####

# Create a function to calculate team form
def calculate_form(team, date, df_filtered, n_matches=5):
    past_matches = df_filtered[(df_filtered['HomeTeam'] == team) | (df_filtered['AwayTeam'] == team)]
    past_matches = past_matches[past_matches['DateTime'] < date].tail(n_matches)
    
    form = 0
    for _, row in past_matches.iterrows():
        if row['HomeTeam'] == team:
            if row['FTR'] == 2:  # Home win
                form += 3
            elif row['FTR'] == 1:  # Draw
                form += 1
        else:
            if row['FTR'] == 0:  # Away win
                form += 3
            elif row['FTR'] == 1:  # Draw
                form += 1
    return form / n_matches  # Average form

# Apply to HomeTeam and AwayTeam
df_filtered['HomeForm'] = df_filtered.apply(lambda x: calculate_form(x['HomeTeam'], x['DateTime'], df_filtered), axis=1)
df_filtered['AwayForm'] = df_filtered.apply(lambda x: calculate_form(x['AwayTeam'], x['DateTime'], df_filtered), axis=1)
df_filtered.tail()
```

```python
# Function to calculate head-to-head win rate
def head_to_head(home_team, away_team, date, df_filtered, n_matches=5):
    past_matches = df_filtered[((df_filtered['HomeTeam'] == home_team) & (df_filtered['AwayTeam'] == away_team)) |
                      ((df_filtered['HomeTeam'] == away_team) & (df_filtered['AwayTeam'] == home_team))]
    past_matches = past_matches[past_matches['DateTime'] < date].tail(n_matches)
    
    home_wins = 0
    for _, row in past_matches.iterrows():
        if row['HomeTeam'] == home_team and row['FTR'] == 2:
            home_wins += 1
        elif row['AwayTeam'] == home_team and row['FTR'] == 0:
            home_wins += 1
    return home_wins / n_matches  # Home team win rate

df_filtered['H2H'] = df_filtered.apply(lambda x: head_to_head(x['HomeTeam'], x['AwayTeam'], x['DateTime'], df_filtered), axis=1)
```

```python
# Calculate rest days between matches for each team
def calculate_rest_days(team, date, df_filtered):
    # Filter past matches for the given team
    past_matches = df_filtered[((df_filtered['HomeTeam'] == team) | 
                                (df_filtered['AwayTeam'] == team)) & 
                               (df_filtered['DateTime'] < date)]
    
    # If no past matches, it's a newly promoted team
    if past_matches.empty:
        return 45  # More realistic rest days for a newly promoted team

    # Get the most recent match date
    last_match_date = past_matches['DateTime'].max()
    
    # Calculate rest days
    rest_days = (date - last_match_date).days

    # Check for a new season
    if 70 <= rest_days < 100:
        return 30  # More realistic break for offseason
    
    # Check for a newly promoted club
    if rest_days >= 100:
        return 45  # Newly promoted team preseason rest
    
    return rest_days

# Apply function row-wise
df_filtered['HomeDaysRest'] = df_filtered.apply(lambda x: calculate_rest_days(x['HomeTeam'], x['DateTime'], df_filtered), axis=1)
df_filtered['AwayDaysRest'] = df_filtered.apply(lambda x: calculate_rest_days(x['AwayTeam'], x['DateTime'], df_filtered), axis=1)
```

```python
# Improved metric for expected goals with adjusted relative performance

# Create empty columns for the new metrics
df_filtered['expected_home_goals'] = 0.0
df_filtered['expected_away_goals'] = 0.0

# Loop through each game
for idx, game in df_filtered.iterrows():
    # Get index of current game to ensure we only look at prior games
    current_idx = df_filtered.index.get_loc(idx)
    if current_idx > 0:
        prior_games = df_filtered.iloc[:current_idx]
    else:
        # No prior games, use league averages
        df_filtered.at[idx, 'expected_home_goals'] = df_filtered['FTHG'].mean()
        df_filtered.at[idx, 'expected_away_goals'] = df_filtered['FTAG'].mean()
        continue
    
    # Get team-specific performance metrics
    home_team = game['HomeTeam']
    away_team = game['AwayTeam']
    
    # Get last 30 games (or as many as available) for each team in each context
    home_team_home_games = prior_games[prior_games['HomeTeam'] == home_team].tail(30)
    home_team_away_games = prior_games[prior_games['AwayTeam'] == home_team].tail(30)
    away_team_home_games = prior_games[prior_games['HomeTeam'] == away_team].tail(30)
    away_team_away_games = prior_games[prior_games['AwayTeam'] == away_team].tail(30)
    
    # Calculate league averages from the same period
    recent_league_games = prior_games.tail(100)  # Use recent league games for comparison
    league_avg_home_goals = recent_league_games['FTHG'].mean()
    league_avg_away_goals = recent_league_games['FTAG'].mean()
    
    # Calculate offensive strength (goals scored relative to league average)
    if len(home_team_home_games) > 5:
        home_attack_strength = home_team_home_games['FTHG'].mean() / league_avg_home_goals
    else:
        # Not enough data, use a more conservative estimate
        all_home_games = prior_games[prior_games['HomeTeam'] == home_team]
        home_attack_strength = 1.0 if len(all_home_games) == 0 else all_home_games['FTHG'].mean() / league_avg_home_goals
    
    if len(away_team_away_games) > 5:
        away_attack_strength = away_team_away_games['FTAG'].mean() / league_avg_away_goals
    else:
        all_away_games = prior_games[prior_games['AwayTeam'] == away_team]
        away_attack_strength = 1.0 if len(all_away_games) == 0 else all_away_games['FTAG'].mean() / league_avg_away_goals
    
    # Calculate defensive weakness (goals conceded relative to league average)
    if len(home_team_home_games) > 5:
        home_defense_weakness = home_team_home_games['FTAG'].mean() / league_avg_away_goals
    else:
        all_home_games = prior_games[prior_games['HomeTeam'] == home_team]
        home_defense_weakness = 1.0 if len(all_home_games) == 0 else all_home_games['FTAG'].mean() / league_avg_away_goals
    
    if len(away_team_away_games) > 5:
        away_defense_weakness = away_team_away_games['FTHG'].mean() / league_avg_home_goals
    else:
        all_away_games = prior_games[prior_games['AwayTeam'] == away_team]
        away_defense_weakness = 1.0 if len(all_away_games) == 0 else all_away_games['FTHG'].mean() / league_avg_home_goals
    
    # Apply regression to the means to reduce extreme values
    home_attack_strength = 0.7 * home_attack_strength + 0.3
    away_attack_strength = 0.7 * away_attack_strength + 0.3
    home_defense_weakness = 0.7 * home_defense_weakness + 0.3
    away_defense_weakness = 0.7 * away_defense_weakness + 0.3
    
    # Calculate expected goals using the Poisson expectation formula
    # Home team expected goals = league avg home goals  home attack strength  away defense weakness
    df_filtered.at[idx, 'expected_home_goals'] = league_avg_home_goals * home_attack_strength * away_defense_weakness
    
    # Away team expected goals = league avg away goals  away attack strength  home defense weakness
    df_filtered.at[idx, 'expected_away_goals'] = league_avg_away_goals * away_attack_strength * home_defense_weakness

# Round the expected goals to 2 decimal places for readability
df_filtered['expected_home_goals'] = df_filtered['expected_home_goals']
df_filtered['expected_away_goals'] = df_filtered['expected_away_goals']
```

```python
# Calculate newly promted teams
# Add NewlyPromoted column (1 if rest days == 45, else 0)
df_filtered['HomeNewlyPromoted'] = df_filtered.apply(lambda row: 1 if row['HomeDaysRest'] == 45 and row['Season'] != '2014/2015' else 0, axis=1)
df_filtered['AwayNewlyPromoted'] = df_filtered.apply(lambda row: 1 if row['AwayDaysRest'] == 45 and row['Season'] != '2014/2015' else 0, axis=1)

```

```python
# Calculate Shooting Efficiency for Home and Away teams

# Avoid division by zero
df_filtered['HomeShootingEfficiency'] = df_filtered.apply(lambda x: x['HST'] / x['HS'] if x['HS'] > 0 else 0, axis=1)
df_filtered['AwayShootingEfficiency'] = df_filtered.apply(lambda x: x['AST'] / x['AS'] if x['AS'] > 0 else 0, axis=1)

# Extract starting year from season (e.g., '2024/2025' -> 2024)
df_filtered['SeasonStartYear'] = df_filtered['Season'].str.split('/').str[0].astype(int)

# Decay factor for older matches
decay_factor = 0.8  

# Initialize new feature columns
df_filtered['HomeAvgShootingEfficiency'] = 0.0
df_filtered['AwayAvgShootingEfficiency'] = 0.0

# Dictionary to store past shooting efficiency per team
team_shooting_eff = {}

# Process each match in chronological order
df_filtered = df_filtered.sort_values(by='DateTime')

for index, row in df_filtered.iterrows():
    home_team, away_team, match_date = row['HomeTeam'], row['AwayTeam'], row['DateTime']

    # Get past matches for both teams before the current match
    past_home_matches = df_filtered[(df_filtered['HomeTeam'] == home_team) | (df_filtered['AwayTeam'] == home_team)]
    past_home_matches = past_home_matches[past_home_matches['DateTime'] < match_date]
    
    past_away_matches = df_filtered[(df_filtered['HomeTeam'] == away_team) | (df_filtered['AwayTeam'] == away_team)]
    past_away_matches = past_away_matches[past_away_matches['DateTime'] < match_date]

    # Compute weighted average shooting efficiency using decay factor
    def compute_weighted_efficiency(past_matches, team):
        if past_matches.empty:
            return 0  # No past matches means no prior efficiency

        past_matches['SeasonStartYear'] = past_matches['Season'].str.split('/').str[0].astype(int)
        past_matches['DecayWeight'] = decay_factor ** (2025 - past_matches['SeasonStartYear'])

        # Compute efficiency based on whether team was home or away
        home_matches = past_matches[past_matches['HomeTeam'] == team]
        away_matches = past_matches[past_matches['AwayTeam'] == team]

        home_eff = (home_matches['HomeShootingEfficiency'] * home_matches['DecayWeight']).sum() / home_matches['DecayWeight'].sum() if not home_matches.empty else 0
        away_eff = (away_matches['AwayShootingEfficiency'] * away_matches['DecayWeight']).sum() / away_matches['DecayWeight'].sum() if not away_matches.empty else 0

        return (home_eff + away_eff) / 2  # Average across home and away matches

    # Assign past efficiency to the current match
    df_filtered.at[index, 'HomeAvgShootingEfficiency'] = compute_weighted_efficiency(past_home_matches, home_team)
    df_filtered.at[index, 'AwayAvgShootingEfficiency'] = compute_weighted_efficiency(past_away_matches, away_team)

# Drop unnecessary columns
df_filtered = df_filtered.drop(columns=['HS', 'AS', 'HST', 'AST', 'HomeShootingEfficiency', 'AwayShootingEfficiency'])

# Display the updated dataframe
df_filtered.head()
```
